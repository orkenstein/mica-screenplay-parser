```python
import spacy
from spacy import displacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
text = "Sir Lewis Carl Davidson Hamilton MBE HonFREng is a British racing driver. He currently competes in Formula One for Mercedes, having previously driven for McLaren from 2007 to 2012."
```

```python
doc = nlp(text)
```

```python
for sent in doc.sents:
    print(sent)
    for token in sent:
        print("\t", token, token.tag_, token.dep_)
        displacy.render(sent, style="dep")
```
